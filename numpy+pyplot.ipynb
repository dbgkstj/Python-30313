```python
t=np.arange(0,12,0.1)
y1=np.sin(t)
y2=np.cos(t)

plt.plot(t,y1, 'r--', marker='*')
plt.plot(t,y2, 'go-')
plt.grid()
plt.figure(figsize=(12,6))
```

```python
plt.subplot(1,2,1)
plt.plot(t,y1, 'r--', marker='*')
plt.subplot(1,2,2)
plt.plot(t,y2, 'go-')
```
